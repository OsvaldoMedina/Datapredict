```python
# Manteniendo la variable  'insulin' y  re vomiendo las variables restates a la medicación de deabetis

df.drop(['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 
           'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 
           'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone', 'tolazamide', 'examide', 
           'citoglipton', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone',
           'metformin-rosiglitazone', 'metformin-pioglitazone'], axis=1, inplace=True)
```

```python
# denote 'diag_1' as '1' if it relates to diabetes and '0' if it's not
# remove 'diag_2' and 'diag_3'
df['diag_1'] = pd.Series([1 if val.startswith('200') else 0 for val in df['diag_1']], index=df.index)
df.drop(['diag_2', 'diag_3'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# we end up with 98052 rows, 20 features, and one response variable 'readmitted' 
list(df.columns.values)
```

```python
# one-hot-encoding on categorical features
# convert nominal values to dummy values

# df_ancur_id = pd.get_dummies(df['encounter_id'])
# df_pati_nbr = pd.get_dummies(df[ 'patient_nbr'])
df_age = pd.get_dummies(df['age'])
df_race = pd.get_dummies(df['race'])
df_gender = pd.get_dummies(df['gender'])
df_max_glu_serum = pd.get_dummies(df['max_glu_serum'])
df_A1Cresult = pd.get_dummies(df['A1Cresult'])
df_insulin = pd.get_dummies(df['insulin'])
df_change = pd.get_dummies(df['change'])
df_diabetesMed = pd.get_dummies(df['diabetesMed'])
df_discharge_disposition_id = pd.get_dummies(df['discharge_disposition_id'])
df_admission_source_id = pd.get_dummies(df['admission_source_id'])
df_admission_type_id = pd.get_dummies(df['admission_type_id'])
```

```python
df = pd.concat([df, df_age, df_race, df_gender, df_max_glu_serum, df_A1Cresult, 
                  df_insulin, df_change, df_diabetesMed, df_discharge_disposition_id, 
                  df_admission_source_id, df_admission_type_id], axis=1)
df.drop(['age', 'race', 'gender', 'max_glu_serum', 'A1Cresult', 'insulin', 'change', 
                  'diabetesMed', 'discharge_disposition_id', 'admission_source_id', 
                  'admission_type_id'], axis=1, inplace=True)
```

```python
# apply square root transformation on right skewed count data to reduce the effects of extreme values.
# here log transformation is not appropriate because the data is Poisson distributed and contains many zero values.
df['number_outpatient'] = df['number_outpatient'].apply(lambda x: np.sqrt(x + 0.5))
df['number_emergency'] = df['number_emergency'].apply(lambda x: np.sqrt(x + 0.5))
df['number_inpatient'] = df['number_inpatient'].apply(lambda x: np.sqrt(x + 0.5))
```

```python
# feature scaling, features are standardized to have zero mean and unit variance
feature_scale_cols = ['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 
                      'number_diagnoses', 'number_inpatient', 'number_emergency', 'number_outpatient']


scaler = preprocessing.StandardScaler().fit(df[feature_scale_cols])
data_scaler = scaler.transform(df[feature_scale_cols])

data_scaler_df = pd.DataFrame(data = data_scaler, columns=feature_scale_cols, index=df.index)
df.drop(feature_scale_cols, axis=1, inplace=True)
df = pd.concat([df, data_scaler_df], axis=1)
```

```python
df = df.dropna()
```

```python
# create X (features) and y (response)
X = df.drop(['readmitted'], axis=1)
y = df['readmitted']
```

```python
# split X and y into cross-validation (75%) and testing (25%) data sets

X_cv, X_test, y_cv, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python

forest = RandomForestClassifier()
forest.fit(X_cv, y_cv)
importances = forest.feature_importances_

# make importance relative to the max importance
feature_importance = 100.0 * (importances / importances.max())
sorted_idx = np.argsort(feature_importance)
feature_names = list(X_cv.columns.values)
feature_names_sort = [feature_names[indice] for indice in sorted_idx]
pos = np.arange(sorted_idx.shape[0]) + .4
print ('Top 6 features are: ')
for feature in feature_names_sort[::-1][:6]:
    print (feature)
```

```python
X_cv_t6 = X_cv[['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_diagnoses',
           'number_inpatient']]
```

```python
# Logistic Regression

clf1 = LogisticRegression(fit_intercept=True)
# LR_score = cross_val_score(clf1, X_cv, y_cv, cv=10, scoring='accuracy').mean()
# LR_score


clf1.fit(X_cv, y_cv)
logit_pred = clf1.predict(X_test)
pd.crosstab(pd.Series(y_test, name = 'Actual'), pd.Series(logit_pred, name = 'Predict'), margins = True)
```

```python
print("Accuracy is {0:.5f}".format(accuracy_score(y_test, logit_pred)))
print("Precision is {0:.5f}".format(precision_score(y_test, logit_pred)))
print("Recall is {0:.5f}".format(recall_score(y_test, logit_pred)))
```

```python
# Naive Bayes

clf2 = GaussianNB()
# NB_score = cross_val_score(clf2, X_cv, y_cv, cv=10, scoring='accuracy').mean()
# NB_score

clf2.fit(X_cv, y_cv)
baye_pred = clf2.predict(X_test)
pd.crosstab(pd.Series(y_test, name = 'Actual'), pd.Series(baye_pred, name = 'Predict'), margins = True)
```

```python
print("Accuracy is {0:.5f}".format(accuracy_score(y_test, baye_pred)))
print("Precision is {0:.5f}".format(precision_score(y_test, baye_pred)))
print("Recall is {0:.5f}".format(recall_score(y_test, baye_pred)))
```

```python
# Random Forest


clf3 = RandomForestClassifier()
# RF_score = cross_val_score(clf1, X_cv, y_cv, cv=10, scoring='accuracy').mean()
# RF_score

clf3.fit(X_cv, y_cv)
rf_pred = clf3.predict(X_test)
pd.crosstab(pd.Series(y_test, name = 'Actual'), pd.Series(rf_pred, name = 'Predict'), margins = True)

```

```python
print("Accuracy is {0:.5f}".format(accuracy_score(y_test, rf_pred)))
print("Precision is {0:.5f}".format(precision_score(y_test, rf_pred)))
print("Recall is {0:.5f}".format(recall_score(y_test, rf_pred)))
```

```python
# Logistic Regression on Top 6 features
# still be able to achieve good result with reduced running time
LR_score_top = cross_val_score(clf1, X_cv_t6, y_cv, cv=10, scoring='accuracy').mean()
LR_score_top
```

```python
# define the parameter values that should be searched
C_range = np.arange(0.1, 3.1, 0.2)
param_grid = dict(C = C_range)
```

```python
# instantiate the grid, fit the grid with data

clf = LogisticRegression()
grid = GridSearchCV(clf, param_grid, cv=10, scoring='accuracy')
grid.fit(X_cv, y_cv)
```

```python
# plot the results

means = grid.cv_results_['mean_test_score']
params = grid.cv_results_['params']
for mean,param in zip(means,params):
    print("%f  with:   %r" % (mean,param))
```

```python
# examine the best model
print (grid.best_score_)
print (grid.best_params_)
print (grid.best_estimator_)
```

```python
# examine the class distribution of the testing set
y_test.value_counts()
```

```python
# calculate null accuracy (for binary classification problems coded as 0/1)
max(y_test.mean(), 1- y_test.mean())
```

```python
# train a logistic regression model on the cross-validation set with the optimal C parameter
logreg = LogisticRegression(C=grid.best_params_['C'])
logreg.fit(X_cv, y_cv)
```

```python

y_pred_class = logreg.predict(X_test)
print (metrics.accuracy_score(y_test, y_pred_class))
```

```python
# Matrix de confusion

confusion = metrics.confusion_matrix(y_test, y_pred_class)
TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]
print (confusion)
```

```python
# Sensitivity 

print (metrics.recall_score(y_test, y_pred_class))

# Specificity: TN/(TN+FP)

print (TN / float(TN + FP))

# False Positive Rate

print (TP / float(TP + FN))

```

```python
# store the predicted probabilities for class 1
y_pred_prob = logreg.predict_proba(X_test)[:, 1]
```

```python
# plot ROC curve

fig = plt.figure(figsize=(10,9))
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for diabetes readmission')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```
